```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
labels = ['Young Adults','Adult','Middle Aged','Senior']                                                  #Creating a new column to categorize the age of the customers
df['age_group']=pd.cut(df['age'],bins=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df.frequency_of_purchases.value_counts()
```

```python
frequency_mapping = {
'Every 3 Months':90,
'Annually':365,
'Quarterly':90,
'Monthly':30,
'Bi-Weekly':14,
'Fortnightly':14,
'Weekly':7
}
```

```python
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)            #fequency_of_purchase is in wordings which is difficult for analyze so it will be converted into numbers
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
((df['discount_applied'] == df['promo_code_used'])).all()              #Both columns are same if promo code is applied then there must have been a discount given
```

```python
df = df.drop(['promo_code_used'],axis=1)                                                              #so lets drop one column
```

```python
import sqlite3
conn = sqlite3.connect("mydatabase.db")   # create SQL database
cursor = conn.cursor()
```

```python
df.to_sql("mytable", conn, if_exists="replace", index=False)
```

```python
query = "SELECT * FROM mytable"
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
    gender,
    SUM(purchase_amount) AS total_revenue
FROM mytable
GROUP BY gender;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT customer_id, purchase_amount
FROM mytable
WHERE discount_applied = 'Yes'
AND purchase_amount > (
    SELECT AVG(purchase_amount) FROM mytable
);
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
    item_purchased,
    round(AVG(review_rating),2) AS avg_rating
FROM mytable
GROUP BY item_purchased
ORDER BY avg_rating DESC
LIMIT 5;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
    shipping_type,
    round(AVG(purchase_amount),2) AS avg_purchase
FROM mytable
WHERE shipping_type IN ('Express','Standard')
GROUP BY shipping_type;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
    subscription_status,
    round(AVG(purchase_amount),2) AS avg_spend,
    SUM(purchase_amount) AS total_revenue
FROM mytable
GROUP BY subscription_status;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
    item_purchased,
    ROUND(COUNT(CASE WHEN discount_applied = 'Yes' THEN 1 END) * 100.0 / COUNT(*),2) AS discount_percentage
FROM mytable
GROUP BY item_purchased
ORDER BY discount_percentage DESC
LIMIT 5;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
    CASE
        WHEN previous_purchases = 0 THEN 'New'
        WHEN previous_purchases BETWEEN 1 AND 5 THEN 'Returning'
        ELSE 'Loyal'
    END AS customer_segment,
    COUNT(*) AS count
FROM mytable
GROUP BY customer_segment;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT *
FROM (
    SELECT
        category,
        item_purchased,
        COUNT(*) AS purchase_count,
        ROW_NUMBER() OVER (PARTITION BY category ORDER BY COUNT(*) DESC) AS rank
    FROM mytable
    GROUP BY category, item_purchased
)
WHERE rank <= 3;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
    subscription_status,
    COUNT(*) AS count
FROM mytable
WHERE previous_purchases > 5
GROUP BY subscription_status;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
     age_group,
    SUM(purchase_amount) AS total_revenue
FROM mytable
GROUP BY age_group
order by total_revenue desc;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
    category,
    AVG(review_rating) AS avg_rating
FROM mytable
GROUP BY category;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
    category,
    SUM(purchase_amount) AS revenue
FROM mytable
GROUP BY category
ORDER BY revenue DESC
LIMIT 1;
"""
result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT
    customer_id,
    SUM(purchase_amount) AS total_spent,
    RANK() OVER (ORDER BY SUM(purchase_amount) DESC) AS rank
FROM mytable
GROUP BY customer_id;
"""
result = pd.read_sql_query(query, conn)
result
```
